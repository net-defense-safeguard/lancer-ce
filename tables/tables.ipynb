## Pilot Melee Weapons

```python
import pandas as pd
import yaml
```

```python
key_column = 'key'
tag_columns = ['tags', 'range', 'damage']
table_names = ['pilot_melee_weapons', 'pilot_ranged_weapons']
table_file_ext = '.tsv'
tag_columns_by_table_name = {table_name: tag_columns for table_name in table_names}
```

```python
def parse_tags(value):
    if type(value) is str:
        parsed_value = yaml.load(f'[{value}]')
        if len(parsed_value) is 1:
            return parsed_value[0]
        return parsed_value
    return None

def parse_tag_columns(df, tag_columns):
    df_cp = df.copy()
    for tag_column in tag_columns:
        df_cp[tag_column] = df_cp[tag_column].apply(parse_tags)
    return df_cp
```

```python
df_by_table_name = {table_name: parse_tag_columns(pd.read_table(f'{table_name}{table_file_ext}').set_index(key_column), tag_columns_by_table_name[table_name]) for table_name in table_names}
```

```python
df = pd.concat(df_by_table_name.values())
```
