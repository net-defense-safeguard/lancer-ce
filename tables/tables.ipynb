```python
import pandas as pd
import yaml
import pyaml
import os
```

```python
key_column = 'key'
tag_columns = ['tags', 'range', 'damage']
table_names = ['pilot_melee_weapons', 'pilot_ranged_weapons']
table_file_ext = '.tsv'
tag_columns_by_table_name = {table_name: tag_columns for table_name in table_names}
```

```python
def parse_tag(tag, default_tag = None):
    if (type(tag) is not str) and (type(tag) is not dict):
        if type(default_tag) is str:
            return {default_tag: tag}
        else:
            return default_tag
    if len(tag) > 1:
        return tag
    elif len(tag) is 1:
        if list(tag.values())[0] is None:
            return list(tag.keys())[0]
        else:
            return tag
    else:
        return None

def parse_tags(value):
    if type(value) is str:
        tags = [parse_tag(tag) for tag in yaml.safe_load(f'[{value}]') if len(tag) > 0]
        tags = [tag for tag in tags if tag is not None]
        return tags
    return []

def parse_tag_columns(df, tag_columns):
    df_cp = df.copy()
    for tag_column in tag_columns:
        df_cp[tag_column] = df_cp[tag_column].apply(parse_tags)
    return df_cp
```

```python
df_by_table_name = {table_name: parse_tag_columns(pd.read_table(f'{table_name}{table_file_ext}').set_index(key_column), tag_columns_by_table_name[table_name]) for table_name in table_names}
```

```python
df = pd.concat(df_by_table_name.values())
```
